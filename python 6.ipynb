### 모듈 불러오기

- import module-name 으로 불러오기 가능
- dir()함수 : 모듈이나 클래스에 포함된 속성 확인
- from import : 모듈에서 특정 변수, 함수, 클래스 등 가져오기
- form import* : 모든 변수, 함수, 클래스 가져오기(권장x 메모리 낭비 심함)
- import as : 모듈에 별칭 지어주기

- 모듈 탐색 우선순위
    1. 현재 작업 디렉토리 (os.getcwd()로 확인 가능)
    2. PYTHONPATH 환경 변수에 지정된 디렉토리
    3. 표준 라이브러리 디렉토리
    4. site-packages 디렉토리

    - add(x, y): 두 숫자 x와 y를 더한 값을 반환합니다.
    - subtract(x, y): x에서 y를 뺀 값을 반환합니다.
    - multiply(x, y): x와 y를 곱한 값을 반환합니다.
    - divide(x, y): x를 y로 나눈 값을 반환합니다. 단, y가 0이면 "0으로 나눌 수 없습니다."라는 메시지를 출력합니다.
- main.py라는 파일을 만들어 mymath 모듈을 임포트하고, 각 함수를 호출하여 결과를 출력하세요.


```python
import mymath

x = 4
y = 3

print(mymath.add(x,y))
print(mymath.subtract(x,y))
print(mymath.multiply(x,y))
print(mymath.divide(x,y))
```
